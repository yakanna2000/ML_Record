```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, KFold, GridSearchCV, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Step 2: Load dataset
df = pd.read_csv('Week8_Iris.csv')  # Replace with your actual path

# Step 3: Print no. of rows and columns
print("3. Shape:", df.shape)

# Step 4: Data types
print("\n4. Data types:\n", df.dtypes)

# Step 5: Attribute names
print("\n5. Column Names:\n", df.columns.tolist())

# Step 6: Top 5 rows
print("\n6. Top 5 rows:\n", df.head())

# Step 7: Bottom 5 rows
print("\n7. Bottom 5 rows:\n", df.tail())

# Step 8: Unique class labels
print("\n8. Unique class labels:", df['Species'].unique())

# Step 9: Number of examples per class and missing values
print("\n9. Class distribution:\n", df['Species'].value_counts())
print("Missing values:\n", df.isnull().sum())

# Step 10: Drop ID column
df.drop(columns=['Id'], inplace=True)

# Step 11: Summary statistics
print("\n11. Summary statistics:\n", df.describe())

# Step 12: Histogram of all input features
df.hist(figsize=(10, 8))
plt.suptitle("12. Histograms of all features")
plt.tight_layout()
plt.show()

# Step 13: Scatter plot - SepalLength vs SepalWidth
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title("13. Sepal Length vs Sepal Width")
plt.show()

# Step 14: Scatter plot using FacetGrid
g = sns.FacetGrid(df, hue="Species")
g.map_dataframe(sns.scatterplot, x='SepalLengthCm', y='SepalWidthCm')
g.add_legend()
plt.suptitle("14. Sepal Scatter Plot by Species")
plt.show()

# Step 15: Boxplot - Petal Width by species
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title("15. Petal Width by Species")
plt.show()

# Step 16: Boxplot - Petal Length by species
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title("16. Petal Length by Species")
plt.show()

# Step 17: Pairplot of all input features
sns.pairplot(df, hue='Species')
plt.suptitle("17. Pairplot of All Features", y=1.02)
plt.show()

# Step 18: Correlation matrix
print("\n18. Correlation matrix:\n", df.corr())

# Step 19: Prepare X and y
X = df.drop(columns='Species')
y = df['Species']

# Step 20: Train/test split (70% train)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Step 21: SVC documentation - check: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

# Step 22–23: Instantiate SVC with cost (C=1)
model = SVC(C=1)

# Step 24: Fit model
model.fit(X_train, y_train)

# Step 25: Predict
y_pred = model.predict(X_test)

# Step 26: Confusion matrix
print("\n26. Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Step 27–28: Other metrics
print("\n28. Classification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Step 29: K-fold CV (k=5)
kfold = KFold(n_splits=5, shuffle=True, random_state=1)
cv_scores = cross_val_score(model, X, y, cv=kfold)
print("\n29. Average CV Accuracy:", cv_scores.mean())

# Step 30–34: GridSearchCV for best C value
param_grid = {'C': [0.1, 1, 10, 100, 1000]}
grid_model = GridSearchCV(SVC(), param_grid, cv=5, return_train_score=True)
grid_model.fit(X_train, y_train)

# Step 34: Print grid results
results = pd.DataFrame(grid_model.cv_results_)
print("\n34. GridSearchCV Results:\n", results[['param_C', 'mean_train_score', 'mean_test_score']])

# Step 35: Plot C vs Scores
plt.plot(param_grid['C'], results['mean_train_score'], label='Train Score', marker='o')
plt.plot(param_grid['C'], results['mean_test_score'], label='Test Score', marker='x')
plt.xscale('log')
plt.xlabel('C value (log scale)')
plt.ylabel('Score')
plt.title("35. GridSearchCV: C vs Score")
plt.legend()
plt.show()

# Step 36: Best score and best C
print("\n36. Best Score:", grid_model.best_score_)
print("Best C value:", grid_model.best_params_['C'])
```

```python
# 1. Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# 2. Load dataset
df = pd.read_csv("Week8_Iris.csv")

# 3. Print number of rows and columns
print("Rows and Columns:", df.shape)

# 4. Print dataframe info including data types
print(df.info())

# 5. Print attribute names
print("Attribute names:", df.columns.tolist())

# 6. Print top 5 rows
print(df.head())

# 7. Print bottom 5 rows
print(df.tail())

# 8. Print the unique class labels
print("Unique class labels:", df['Species'].unique())

# 9. Print number of examples from each species; check for missing values
print(df['Species'].value_counts())
print("Missing values:\n", df.isnull().sum())

# 10. Drop the 'Id' column
df = df.drop(columns=['Id'])

# 11. Print summary statistics
print(df.describe())

# 12. Plot histogram of all input features
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# 13. Scatter plot of SepalLengthCm vs SepalWidthCm
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'])
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Width")
plt.show()

# 14. Scatter plot using seaborn’s FacetGrid
sns.lmplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df, fit_reg=False)
plt.title("Sepal Length vs Width by Species")
plt.show()

# 15. Boxplot of PetalWidthCm by Species
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title("Boxplot of Petal Width by Species")
plt.show()

# 16. Boxplot of PetalLengthCm by Species
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title("Boxplot of Petal Length by Species")
plt.show()

# 17. Pairplot of all features colored by Species
sns.pairplot(df, hue='Species')
plt.show()

# 18. Print correlation matrix
print("Correlation matrix:\n", df.drop(columns=['Id', 'Species'], errors='ignore').corr())



# 19. Prepare X and y
X = df.drop(columns=['Species'])
y = df['Species']

# 20. Split the data into 70% train and 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# 21. Read the documentation of sklearn.svm.SVC (done offline)

# 22. Model building step begins

# 23. Instantiate an object of SVC with cost C = 1
svc = SVC(C=1)

# 24. Fit the SVC model
svc.fit(X_train, y_train)

# 25. Predict using the trained model
y_pred = svc.predict(X_test)

# 26. Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# 27. Print additional metrics
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# 28. Already printed above

# 29. K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# 30. Use GridSearchCV to find the optimal value of C
param_grid = {'C': [0.1, 1, 10, 100, 1000]}
grid = GridSearchCV(SVC(), param_grid, cv=kf, return_train_score=True)

# 31 & 32. Already set above

# 33. Fit the model using GridSearchCV
grid.fit(X_train, y_train)

# 34. Print the scores of grid search as dataframe
cv_results = pd.DataFrame(grid.cv_results_)
print(cv_results[['param_C', 'mean_train_score', 'mean_test_score']])

# 35. Plot C vs train/test scores
plt.plot(param_grid['C'], cv_results['mean_train_score'], label='Train Score')
plt.plot(param_grid['C'], cv_results['mean_test_score'], label='Test Score')
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Score')
plt.title('C vs Train/Test Scores')
plt.legend()
plt.show()


# 36. Best score and best C value
print("Best Score:", grid.best_score_)
print("Best C Value:", grid.best_params_['C'])
```
